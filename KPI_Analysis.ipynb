```python
# Select June 2022 onwards
df_filt = df[df['Date of Reporting'] >= '2022-06-01']
```

```python
df_filt.groupby(by='Month-Year of Reporting')['CTR','CPC','Impressions','Clicks','Visits to homepage','Visits to admissions page'].sum()
```

```python
df_agg = df_filt.groupby(by='Month-Year of Reporting')[['CTR','CPC','Impressions','Clicks','Visits to homepage','Visits to admissions page']].sum().sort_values(by='Month-Year of Reporting')
df_agg = df_agg[5:]
df_agg_YoY_Change = round(df_agg.pct_change()*100,2)

df_agg_YoY_Change[1::2]
```

```python
agg_df = df.groupby(by='Month-Year of Reporting')[['CTR','CPC','Impressions','Clicks','Visits to homepage','Visits to admissions page']].sum().sort_values(by='Month-Year of Reporting')
agg_df = agg_df.sort_values(by = 'Month-Year of Reporting')
df_agg = df_agg[5:]
df_agg_YoY_Change = round(df_agg.pct_change()*100,2)
df_agg_YoY_Change[1::2]
```

```python
res_df = round(agg_df.pct_change()*100,2)
res_df
```
